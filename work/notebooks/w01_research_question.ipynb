## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv('../../data/raw/content_refresh_anonymized.csv')
print(f"rows: {len(df):,}  columns: {df.shape[1]}  clients: {df['client_id'].nunique()}")
```

```python
# Only look at pages that are actually visible in search (enough impressions, known position)
visible = df[(df['impressions_90d'] >= 500) & (df['avg_position'] > 0) & (df['avg_position'] <= 20)].copy()
print(f"Visible pages I can compare fairly: {len(visible):,} out of {len(df):,} total rows")

# Compare click-through rate (CTR) by position tier -- position matters a lot
tier_stats = visible.groupby('position_tier')['ctr'].agg(['count', 'median']).sort_values('median')
print(tier_stats)
```

## 4. Careful words: what I can and can't claim

*Write what your work will be able to say (observed, directional, decision-support) — and what it never will (causal proof, 'predicting Google').*

I can claim that these pages get fewer clicks than similar pages in the same 
position. That is something I actually see in the data.

I can say that pages with a bigger gap between their clicks and similar 
pages' clicks are probably worth checking first.

I can't claim that a bad title caused the low clicks. There could be 
other reasons, consolidation (another page on the same site might be 
getting the clicks instead of this one), seasonality (people search 
for this topic less at certain times of year), or a change in how 
Google shows the search results page itself, not the title at all.

I cannot say that fixing the title will definitely bring back more clicks. 
That needs a real test, not just looking at old data.

I cannot say anything about how Google's search algorithm works, since 
I only have click and position numbers, not the algorithm itself.
